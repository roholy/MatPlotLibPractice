```
# import numpy as np #Numpy provides fast mathematical function processing
import pandas as pd # Pandas is used for easy indexing adn data collection tools
# import matplotlib as mpl
# import matplotlib.pyplot as plt
# import seaborn as sns #import library to create heatmap
# %matplotlib inline
company_sales=pd.read_csv("https://raw.githubusercontent.com/roholy/MatPlotLibPractice/main/company_sales_data%20(1).csv")
#this reads from the online github database
company_sales #this shows the database
```

```
%matplotlib inline #import matplot lib to be used as plt
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
```

```
fig = plt.figure()#sets var fig to plot the figure
ax = plt.axes() #sets plt axes to a var ax
plt.plot(company_sales.month_number, company_sales.total_profit) # plot(x,y)
ax.set(xlim=(1,12), ylim=(150000, 4500000),
       xlabel = 'Month Number', ylabel='Total profit',
       title= 'Total Profit of All months');
```

```
ax = plt.axes(label="2")
plt.plot(company_sales.month_number, company_sales.total_profit, linestyle = 'dotted', 
         color='red', label='total Profit Over Time', 
         marker ="o", linewidth = 3, markerfacecolor='red')
plt.legend(loc= 'lower right')
ax.set(xlim=(1,12), ylim=(150000,450000),
      xlabel='Month Number', ylabel='Sold units number',
       title='Total Profit of All Months ');
```

```
ax = plt.axes(label='3')
plt.plot(company_sales.month_number, company_sales.facecream, linestyle='dotted', color='black', label='Facecream')
plt.plot(company_sales.month_number, company_sales.facewash, linestyle='dashed', color='blue', label='Facewash')
plt.plot(company_sales.month_number, company_sales.toothpaste, linestyle='solid', color='red', label='Toothpaste')
plt.plot(company_sales.month_number, company_sales.bathingsoap, linestyle='dashdot', color='green', label='Bathing Soap')
plt.plot(company_sales.month_number, company_sales.shampoo, linestyle='solid', color='purple', label='Shampoo')
plt.plot(company_sales.month_number, company_sales.moisturizer, linestyle='dotted', color='orange', label='Mousturizer')
plt.legend(loc='upper left')
ax.set(xlim=(1,12), ylim=(1,14000),
      xlabel='Month Number', ylabel='Sold units number',
      title='All Product Sales ');
```

```
ax = plt.axes(label='4')
plt.scatter(company_sales.month_number, company_sales.toothpaste, marker='o')
ax.grid(linestyle = '-', linewidth = '0.5', color = 'gray' )
ax.set(xlim=(1,12), ylim=(4000,10000),
      xlabel='Month Number', ylabel='Toothpaste sold units number',
      title='Excercise 4');
```

```
ax = plt.axes(label='5')
plt.bar(company_sales.month_number, company_sales.facewash, color="blue", label='Facewash', width = 0.25)
plt.bar(company_sales.month_number + .25, company_sales.facecream, color = 'yellow', label='Facecream',width =0.25)
plt.legend(loc= 'upper left')
ax.set(xlabel='Month Number', ylabel='Number of Units Sold', title = 'Excercise 5');
```
